## Dependencies

```python
import pandas as pd
from statsmodels import datasets
from matplotlib import pyplot as plt
import seaborn as sns
```

Set the graphing defaults.
(This is optional.)

```python
sns.set_theme()
plt.rc("figure", figsize=(8, 6))
plt.rc("font", size=14)
```

## Notes

- Data was sourced from the Rdatasets
   - Arel-Bundock V (2022). Rdatasets: A collection of datasets originally distributed in various R packages. R package version 1.0.0, https://vincentarelbundock.github.io/Rdatasets.
- `Dataset` is a class from StatsModels
   - See [The Datasets Package](https://www.statsmodels.org/devel/datasets/index.html)
- When initialising the `Series`
   - pass a NumPy `array` to the `data` argument
   - pass an actual argument to the `name` argument
     - this is useful when plotting the time series
   - annual or monthly time series: set a `DatetimeIndex` with `date_range`
     - See [pandas.date_range](https://pandas.pydata.org/docs/reference/api/pandas.date_range.html)
   - quarterly time series: set a `PeriodIndex` with `period_range`
     - See [pandas.period_range](https://pandas.pydata.org/docs/reference/api/pandas.period_range.html)
- Workflow
   1. Load the data
   2. Identify the date of the initial observation
   3. Initialise the `Series`
   4. Plot the time series

## Inputs/Outputs

### Inputs

- *bomregions* (`Dataset`), Australian and Related Historical Annual Climate Data, by region
- *elecequip* (`Dataset`), Electrical equipment manufactured in the Euro area
- *macrodat* (`Dataset`), Macroeconomic Time Series for the United States

### Outputs

- *ts south rain* (`Series`), column *southRain* from input *bomregions* with annual `DatetimeIndex`
- *ts elecequip* (`Series`), input *elecequip* with monthly `DatetimeIndex`
- *ts macrodat* (`Series`), input *macrodat* with quarterly `PeriodIndex`

## Annual time series

### Load the annual data

```python
bomregions = datasets.get_rdataset('bomregions', package='DAAG', cache=True)
bomregions.data.info()
```

### Identify the initial year

```python
bomregions.data['Year'].head(1)
```

### Initialise the annual Series

```python
ts_south_rain = pd.Series(
    data=bomregions.data['southRain'].to_numpy(),
    name='SOUTHERN RAINFALL',
    index=pd.date_range(
        start='1900',
        periods=bomregions.data['southRain'].size,
        freq='A-DEC',
        name='YEAR'
    )
)
ts_south_rain.info()
```

### Plot the annual time series

```python
ts_south_rain.plot(lw=2, label='obs')
plt.ylabel(ts_south_rain.name)
plt.legend()
plt.show()
```

## Monthly time series

### Load the monthly data

```python
elecequip = datasets.get_rdataset('elecequip', package='fpp2', cache=True)
elecequip.data.info()
```

### Identify the initial month

```python
elecequip.data['time'].head(1)
```

### Initialise the monthly Series

```python
ts_elecequip = pd.Series(
    data=elecequip.data['value'].to_numpy(),
    name='NEW ORDERS INDEX',
    index=pd.date_range(
        start='1996-01',
        periods=elecequip.data['value'].size,
        freq='M',
        name='MONTH'
    )
)
ts_elecequip.info()
```

### Plot the monthly time series

```python
ts_elecequip.plot(lw=2, label='obs')
plt.ylabel(ts_elecequip.name)
plt.legend()
plt.show()
```

## Quarterly time series

### Load the quarterly data

```python
macrodat = datasets.get_rdataset('Macrodat', package='Ecdat', cache=True)
macrodat.data.info()
```

### Identify the initial quarter

According to the [documentation](https://vincentarelbundock.github.io/Rdatasets/doc/Ecdat/Macrodat.html), the initial quarter is the first quarter of 1959.

### Initialise the quarterly Series

Note that we use a `period_range`, not a `date_range`.

```python
ts_exruk = pd.Series(
    data=macrodat.data['exruk'].to_numpy(),
    name='USD-GBP EXCHANGE RATE',
    index=pd.period_range(
        start='1959-01-01',
        periods=macrodat.data['exruk'].size,
        freq='Q',
        name='QUARTER'
    )
)
ts_exruk.info()
```

### Plot the quarterly time series

```python
ts_exruk.plot(lw=2, label='obs')
plt.ylabel(ts_exruk.name)
plt.legend()
plt.show()
```

# Hypothesis Testing: Two Sample Z-Test of Population Means

## Sumary

Perform a two sample *z*-test population means.

Data on the mean pass rate across all UK test centres during the period from April 2014 to March 2015 was obtained and analysed using an approximate normal model.

We used a two sample *z*-test to test the null hypothesis that the overall mean pass rate for males was equal to that of females.

## Dependencies

```python
import pandas as pd
from scipy import stats as st
from statsmodels.stats import weightstats as ws
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

## Constants

```python
URL = ('https://raw.githubusercontent.com/ljk233/laughingrook-datasets'
       + '/main/uk_prac_driving_tests/pass_rates.csv')
```

## Main

### Load data

```python
pass_rates = pd.read_csv(URL)
```

```python
pass_rates = pass_rates[['female', 'male']]
```

### Describe the data

```python
pass_rates.info()
```

```python
pass_rates.describe().T
```

```python
pd.DataFrame(
    data=[ws.zconfint(x1=pass_rates['female']),
          ws.zconfint(x1=pass_rates['male'])],
    columns=['lcb', 'ucb'],
    index=['female', 'male']
)
```

```python
_g = sns.displot(
        data=pass_rates.melt(),
        x='value',
        kind='hist',
        col='variable',
        hue='variable',
        legend=False,
        kde=True,
        stat='density'
)
```

```python
_f, _axs = plt.subplots(1, 2, figsize=(11.8, 6), sharey=True)
st.probplot(x=pass_rates['female'], plot=_axs[0])
st.probplot(x=pass_rates['male'], plot=_axs[1])
_f.suptitle('Probability Plots', fontsize=16)
_axs[0].set_title('female')
_axs[1].set_title('male')
plt.show()
```

### Perform the *z*-test

```python
_r = ws.ztest(x1=pass_rates['female'], x2=pass_rates['male'])
pd.Series(data={'zstat': _r[0], 'pvalue': _r[1]}).round(6)
```

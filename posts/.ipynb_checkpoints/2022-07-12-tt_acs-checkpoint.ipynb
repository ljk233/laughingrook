We followed [@thomas_mock](https://twitter.com/thomas_mock)'s' advice and plotted a selected set of graphs that would be useful in an exploratory data analysis (EDA).

Source data was taken from [census.gov](https://www.census.gov/) (via [kaggle.com](kaggle.com).

The data was tidy.
The columns were in PascalCase, so these were transformed to snake_case.
A new multi-index was constructed for the data as `(state, county, census_id,)`, as these were unique identifiers.
Any missing values were replaced with the column mean.

The data was further processed.
We calculated the proportion of each county's population that were women, and rescaled the income per cap by dividing it by 1000.

Given the size of the data, we selected a subset of the variables to explore.
These were:

- women (%age of population)
- poverty
- public work
- income per cap
  - This was rescaled by a factor of 1000

We also sampled the data (*n* = 1250) to get around Altair's max row limit for in-notebook DataFrames.

Four visualisations were produced:
The first shows a **multiple boxplot.**
The second and third are **correlation and covariance heatmaps** (of the standardised data.)
And the fourth and final is a **matrix scatterplot.**

## Dependencies

```python
#| code-fold: false
import pandas as pd
import altair as alt
from matplotlib import pyplot as plt
import seaborn as sns
```

## Constants

```python
#| code-fold: false
ACS_URL = ('https://raw.githubusercontent.com/rfordatascience/tidytuesday/'
           + 'master/data/2018/2018-04-30/week5_acs2015_county_data.csv')
```

## Functions

```python
#| code-fold: false
def pascal_to_snake_case(s: str) -> str:
    return ''.join(['_' + c.lower() if c.isupper() else c for c in s])[1:]
```

```python
#| code-fold: false
def standardise(df: pd.DataFrame) -> pd.DataFrame:
    """@pipeline.
    Standardise the dataframe.
    """
    df = df.transform(lambda x: (x - x.mean()) / x.std())
    return df
```

## Main

### Load the data

```python
acs = pd.read_csv(ACS_URL, encoding='latin-1')
```

### Tidy the data

Rename the columns from `PascalCase` to `snake_case`, and set a mulit-level index.

```python
v_acs = (
    acs
    .rename(columns={s: pascal_to_snake_case(s) for s in acs.columns})
    .set_index(['state', 'county', 'census_id'])
)
```

### Expand the data

Calculate the proportion of the popuation that are females and rescale the *income per cap.*

```python
v_acs['prop_women'] = (
    v_acs['women']
    .div(v_acs[['men', 'women']].sum(axis=1))
    .mul(100)
    .round(1)
)
v_acs['scaled_income_per_cap'] = v_acs['income_per_cap'].div(1000).round(2)
```

### Select and sample the data

```python
v_sample = (
    v_acs
    .get(['scaled_income_per_cap', 'prop_women', 'poverty', 'public_work'])
    .sample(n=1250, random_state=20180430)
)
```

### Standardise the sampled data

This is for the correlation and covariance heatmaps, where we are more interested in the spread of the variables relative to each other, rather than then observed values.

```python
zv_sample = v_sample.transform(standardise)
```

### Visualise the data

```python
alt.Chart(v_sample.melt()).mark_boxplot(size=50).encode(
    x='variable',
    y='value',
    color=alt.Color('variable', legend=None)
).properties(
    height=400,
    width=600,
    title='Multiple boxplots of the selected ACS data'
)
```

```python
corr_df = (
    zv_sample
    .corr()
    .reset_index(drop=False)
    .rename(columns={'index': 'X1'})
    .melt(id_vars='X1', var_name='X2', value_name='r')
    .round(2)
)

base_r = alt.Chart(corr_df).encode(
    x='X1',
    y='X2'
)

heatmap = base_r.mark_rect().encode(
    color=alt.Color("r", scale=alt.Scale(scheme="turbo"))
).properties(
)

text = base_r.mark_text(baseline='middle', size=16).encode(
    text='r',
    color=alt.condition(
        abs(alt.datum.r) > 0.5,
        alt.value('white'),
        alt.value('black')
    )
)

(heatmap + text).properties(
    title='Correlation heatmap',
    width=400,
    height=400,
)
```

```python
cov_df = (
    zv_sample
    .cov()
    .reset_index(drop=False)
    .rename(columns={'index': 'X1'})
    .melt(id_vars='X1', var_name='X2', value_name='cov')
    .round(2)
)

base_cov = alt.Chart(cov_df).encode(
    x='X1',
    y='X2'
)

heatmap = base_cov.mark_rect().encode(
    color=alt.Color("cov", scale=alt.Scale(scheme="turbo"))
)

text = base_cov.mark_text(baseline='middle', size=16).encode(
    text='cov',
    color=alt.condition(
        abs(alt.datum.cov) > 0.5,
        alt.value('white'),
        alt.value('black')
    )
)

(heatmap + text).properties(
    width=400,
    height=400,
    title='Covariance heatmap'
)
```

```python
sns.pairplot(
    v_sample,
    diag_kws={'bins': 20},
    height=1.9
)
plt.show()
```
